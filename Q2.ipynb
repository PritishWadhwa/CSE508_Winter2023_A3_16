```python
file = open("./Data/p2p-Gnutella08.txt", "r")
graph = file.read().splitlines()[4:]
file.close()
```

```python
newGraph = [i.split("\t") for i in graph]

newGraph = [[int(j) for j in i] for i in newGraph]
```

```python
# This is a list of nodes in the Graph.
# Used for the adjacency matrix
#  And certain other functions
listOfNodes = set()
for i in range(len(newGraph)):
    listOfNodes.add(newGraph[i][0])
    listOfNodes.add(newGraph[i][1])
```

# Question 2

## Form the Graph

```python
# 1. Form the graph
g = nx.DiGraph()
# 2. Add the edges to the graph
g.add_edges_from(newGraph)
```

## Page Rank

```python
# 3. Get the Page Rank of the nodes
pr = nx.pagerank(g) 
# 4. Sort the nodes by their Page Rank
sorted_pr = sorted(pr.items(), key=lambda x: x[1], reverse=True)
# 5. Print the nodes
print("Node\t Page Rank")
for node, rank in sorted_pr:
    print(node,"\t", rank)
```

```python
# Get the hubs and authority scores
hubScore, authority = nx.hits(g)
```

## Authority of each node

```python
# Sort the nodes by their authority
sorted_auth = sorted(authority.items(), key=lambda x: x[1], reverse=True)
# Print the nodes
print("Node \t Authority")
for node, auth in sorted_auth:
    print(node,"\t", auth)
```

## Hub Score for node

```python
# Sort the nodes by their hub score
sorted_hub = sorted(hubScore.items(), key=lambda x: x[1], reverse=True)
# Print the nodes
print("Node \t Hub Score")
for node, hub in sorted_hub:
    print(node,"\t", hub)
```

## Comparision of Page Rank with Authority and Hub Score.

```python
import matplotlib.pyplot as plt
nodeList = list(listOfNodes)
nodeList.sort()
# Get the hub score, authority and page rank of the nodes
hubScoreList = [hubScore[i] for i in nodeList]
authorityList = [authority[i] for i in nodeList]
pageRankList = [pr[i] for i in nodeList]
```

```python
n = 10
plt.title("Scores vs Nodes [" + str(n) + " nodes]")
plt.plot(nodeList[:n],hubScoreList[:n], label="Hub Score", color="red")
plt.plot(nodeList[:n],authorityList[:n], label="Authority", color="green")
plt.plot(nodeList[:n],pageRankList[:n], label="Page Rank", color="blue")
plt.xlabel("Nodes")
plt.ylabel("Scores")
plt.legend()
plt.show()
```

```python
n = 30
plt.title("Scores vs Nodes [" + str(n) + " nodes]")
plt.plot(nodeList[:n],hubScoreList[:n], label="Hub Score", color="red")
plt.plot(nodeList[:n],authorityList[:n], label="Authority", color="green")
plt.plot(nodeList[:n],pageRankList[:n], label="Page Rank", color="blue")
plt.xlabel("Nodes")
plt.ylabel("Scores")
plt.legend()
plt.show()
```

```python
nodes = len(listOfNodes)
inDegreeList = dict(zip(list(listOfNodes), [0]*nodes))
outDegreeList = dict(zip(list(listOfNodes), [0]*nodes))
for i in newGraph:
    fromNode = i[0]
    toNode = i[1]
    inDegreeList[toNode] += 1
    outDegreeList[fromNode] += 1
```

```python
n = 10
print("Nodes \t PageRank")
for i in range(n):
    print(sorted_pr[i][0], "\t", sorted_pr[i][1])
print("In Degree",inDegreeList[sorted_pr[0][0]])
print("Out Degree",outDegreeList[sorted_pr[0][0]])

```

```python
print("Nodes \t Hub Score")
for i in range(n):
    print(sorted_hub[i][0], "\t" , sorted_hub[i][1])
print("In Degree",inDegreeList[sorted_hub[0][0]])
print("Out Degree",outDegreeList[sorted_hub[0][0]])
```

```python
print("Nodes \t Authority")
for i in range(n):
    print(sorted_auth[i][0], "\t" ,sorted_auth[i][1])
print("In Degree",inDegreeList[sorted_auth[0][0]])
print("Out Degree",outDegreeList[sorted_auth[0][0]])
```

```python
n = 10
# Take first element of every tuple
top_n_pr = [i[0] for i in sorted_pr[:n]]
top_n_hubs = [i[0] for i in sorted_hub[:n]]
top_n_auth = [i[0] for i in sorted_auth[:n]]

top_n = list(set(top_n_pr).union(set(top_n_hubs)).union(set(top_n_auth)))
hubScoreList = [hubScore[i] for i in top_n]
authorityList = [authority[i] for i in top_n]
pageRankList = [pr[i] for i in top_n]


plt.title("Nodes with Top Scores")
plt.scatter(top_n,hubScoreList, label="Hub Score", color="red")
plt.scatter(top_n,authorityList, label="Authority", color="green")
plt.scatter(top_n,pageRankList, label="Page Rank", color="blue")
plt.xlabel("Nodes")
plt.ylabel("Scores")
plt.legend()
plt.show()
```
