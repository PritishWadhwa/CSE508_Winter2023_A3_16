```python
file = open("./Data/p2p-Gnutella08.txt", "r")
graph = file.read().splitlines()[4:]
file.close()
```

```python
newGraph = [i.split("\t") for i in graph]

newGraph = [[int(j) for j in i] for i in newGraph]
```

# Question 2

## Form the Graph

```python
# 1. Form the graph
g = nx.DiGraph()
# 2. Add the edges to the graph
g.add_edges_from(newGraph)
```

## Page Rank

```python
# 3. Get the Page Rank of the nodes
pr = nx.pagerank(g) 
# 4. Sort the nodes by their Page Rank
sorted_pr = sorted(pr.items(), key=lambda x: x[1], reverse=True)
#.  5Print the nodes
print("Node\t Page Rank")
for node, rank in sorted_pr:
    print(node,"\t", rank)
```

```python
# Get the hubs and authority scores
hubs, auth = nx.hits(g)
```

## Authority of each node

```python
# Sort the nodes by their authority
sorted_auth = sorted(auth.items(), key=lambda x: x[1], reverse=True)
# Print the nodes
print("Node\t Authority")
for node, rank in sorted_auth:
    print(node,"\t", rank)
```

## Hub Score for node

```python
# Sort the nodes by their hub
sorted_hub = sorted(hubs.items(), key=lambda x: x[1], reverse=True)
# Print the nodes
print("Node\t Hub")
for node, rank in sorted_hub:
    print(node,"\t", rank)
```
